# MetaKernel Python Demo

This notebook demonstrates the magics available in [MetaKernel](https://github.com/Calysto/metakernel).

The kernel running this notebook is **MetaKernel Python** — a MetaKernel subclass that directly evaluates Python code using Python's `eval`/`exec`. It provides the full MetaKernel magic system on top of a lightweight Python interpreter.

MetaKernel provides a rich set of line and cell magics that work across any kernel built on it.

## List Available Magics

Use `%lsmagic` to see all installed line and cell magics.

```python
%lsmagic
```

## Shell Magics

`%shell` runs a single shell command. `%%shell` runs the entire cell as a shell script.
You can also use `!` and `!!` as shorthand.

```python
%shell echo "Hello from the shell!"
```

```python
%%shell
echo "Current directory: $(pwd)"
echo "Python version: $(python --version)"
```

## File System Magics

`%ls` lists directory contents. `%cd` changes the working directory.

```python
%ls .
```

## Writing and Running Files

`%%file` writes the cell contents to a file. `%run` executes a file.

```python
%%file /tmp/hello.py
def greet(name):
    return f"Hello, {name}!"

print(greet("MetaKernel"))
```

```python
%run /tmp/hello.py
```

## Timing Magic

`%%time` reports how long a cell takes to execute.

```python
%%time
total = sum(range(1_000_000))
print(total)
```

## HTML Magic

`%html` and `%%html` render output as HTML in the browser.

```python
%html <b>Bold</b>, <i>italic</i>, and <span style="color:steelblue">colored</span> text.
```

```python
%%html
<table border="1" style="border-collapse:collapse; padding:6px">
  <tr><th>Magic</th><th>Description</th></tr>
  <tr><td>%lsmagic</td><td>List all magics</td></tr>
  <tr><td>%shell</td><td>Run a shell command</td></tr>
  <tr><td>%%file</td><td>Write cell to file</td></tr>
  <tr><td>%%time</td><td>Time cell execution</td></tr>
  <tr><td>%%latex</td><td>Render as LaTeX</td></tr>
</table>
```

## LaTeX Magic

`%latex` and `%%latex` render mathematical notation using LaTeX.

```python
%latex $E = mc^2$
```

```python
%%latex
$$\int_{-\infty}^{\infty} e^{-x^2} dx = \sqrt{\pi}$$

$$F(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}(x-a)^n$$
```

## Set and Get Variables

`%set` assigns a Python value to a kernel variable. `%get` retrieves it.

```python
%set x 42
%set items [1, 2, 3, 4, 5]
```

```python
%get x
```

```python
%get items
```

## Macro Magic

`%%macro` defines a reusable named macro. `%macro NAME` executes it.

```python
%%macro hello
print("Hello from a macro!")
print("The answer is", 6 * 7)
```

```python
%macro hello
```

## Tutor Magic

`%%tutor` visualizes code execution step-by-step using [Python Tutor](https://pythontutor.com).

```python
%%tutor --language=python3
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

result = factorial(5)
print(result)
```
